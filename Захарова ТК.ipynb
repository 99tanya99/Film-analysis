```python
columns = ['runtime_clear', 'gross_earn_clear']
new_dataset = dataset.copy()
for col in columns:
    limiter = new_dataset[col].quantile(0.99)
    new_dataset = new_dataset[new_dataset[col] <= limiter]
```

```python
g = sns.PairGrid(dataset, vars=columns)
g.map(sns.scatterplot, alpha=0.5)
plt.suptitle('Диаграммы рассеяния для runtime_clear и gross_earn_clear', y=1.02)
plt.show()
```

```python
g = sns.PairGrid(new_dataset, vars=columns)
g.map(sns.scatterplot, alpha=0.5)
plt.suptitle('Диаграммы рассеяния для runtime_clear и gross_earn_clear', y=1.02)
plt.show()
```
